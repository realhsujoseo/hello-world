```python
#Ans1
import pandas as pd
import numpy as np
df = pd.read_csv('/content/drive/MyDrive/DATA/ramen-ratings.csv',index_col = 0)
df.head()
```

```python
#Ans2
New_df= df.rename(columns={'Top Ten': 'Top10', 'Stars':'Score'})
New_df.head()
```

```python
New_df
```

```python
#Ans3
New_df.loc[:,("Brand","Score")]
```

```python
#An4
New_df["Top10"].fillna("NaN", inplace = True)
New_df.loc[:,"Top10"]
```

```python
print(New_df.loc[(New_df["Top10"] != "NaN")].shape)
New_df.loc[(New_df["Top10"] != "NaN")]
```

```python
import re
New_df["Top10"] = New_df["Top10"].str.replace(r'\n','NaN')
print(New_df.loc[(New_df["Top10"] != "NaN")].shape)
New_df.loc[(New_df["Top10"] != "NaN")]
```

```python
#Ans5
New_df['Score'] = pd.to_numeric(New_df['Score'],errors = 'coerce')
Nissin_high_score = New_df.loc[(New_df['Brand']== 'Nissin') & (New_df['Score']>3.5)]
Nissin_high_score.head()
```

```python
#Ans6
groupby_df = New_df.groupby('Brand').Score.agg([min,max,np.mean])
groupby_df.head()
```

```python
#Ans7
New_df['Score'].mode()
```

```python
#Ans8
groupby_df2 =  New_df.groupby(['Brand','Variety']).Score.agg([np.mean])
groupby_df2
```

```python
groupby_df2.sort_values('mean', ascending= False)
#Mi Goreng Jumbo Beef from Indomie has the highest average.
```
